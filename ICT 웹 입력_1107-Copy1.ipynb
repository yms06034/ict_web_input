```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup as BS
from datetime import datetime

import pandas as pd
import time
import sys, os
import re
import json
import random, math


def find_css(css_selector, browser):
    return browser.find_element(By.CSS_SELECTOR, css_selector)
def finds_css(css_selector, browser):
    return browser.find_elements(By.CSS_SELECTOR, css_selector)

def find_xpath(xpath, browser):
    return browser.find_element(By.XPATH, xpath)
def finds_xpath(xpath, browser):
    return browser.find_elements(By.XPATH, xpath)

def find_id(e_id, browser):
    return browser.find_element(By.ID, e_id)

def find_className(cn, browser):
    return browser.find_element(By.CLASS_NAME, cn)
def finds_className(cn , browser):
    return browser.find_elements(By.CLASS_NAME, cn)

def find_linktext(lt, browser):
    return browser.find_element(By.LINK_TEXT, lt)

def find_name(name, browser):
    return browser.find_element(By.NAME, name)
def finds_name(name, browser):
    return browser.find_elements(By.NAME, name)

def find_tagName(tag_name, browser):
    return browser.find_element(By.TAG_NAME, tag_name)

def finds_tagName(tag_name, browser):
    return browser.find_elements(By.TAG_NAME, tag_name)

def resource_path(relative_path):
    """ Get absolute path to resource, works for dev and for PyInstaller """
    base_path = getattr(sys, '_MEIPASS', os.path.dirname(os.path.abspath(__file__)))
    return os.path.join(base_path, relative_path)
```

```python
def open_browser():
    options = webdriver.ChromeOptions()
    options.add_argument('--no--sandbox')
    options.add_argument('no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--window-size=1080,800')
    options.add_argument('incognito')
    chrome_service = Service('chromedriver')
    chrome_service = Service(executable_path="chromedriver.exe")
    browser = webdriver.Chrome(service=chrome_service, options=options)
    
    return browser
```

```python
def first_page_field(browser, name, part, pos, tel, email, bn, crn):
    # 사업자 번호
    bn_data = find_name('conum3', browser).get_attribute('value')

    if bn_data == '':
        if bn == '0':
            find_name('conum1', browser).send_keys('000')
            time.sleep(.5)
            find_name('conum2', browser).send_keys('00')
            time.sleep(.5)
            find_name('conum3', browser).send_keys('00000')
            time.sleep(.5)
        else:
            find_name('conum1', browser).send_keys(bn[:3])
            time.sleep(.5)
            find_name('conum2', browser).send_keys(bn[3:5])
            time.sleep(.5)
            find_name('conum3', browser).send_keys(bn[5:])
            time.sleep(.5)

    # 법인등록 번호
    crn_data = find_name('conum5', browser).get_attribute('value')
    if crn_data == '':
        if crn != '0':
            find_name('conum4', browser).send_keys(crn[:6])
            time.sleep(.5)
            find_name('conum5', browser).send_keys(crn[6:])
            time.sleep(.5)

    find_name('name', browser).send_keys(name)
    time.sleep(.5)

    if part != 0:
        find_name('part', browser).send_keys(part)
        time.sleep(.5)
    if pos != 0:
        find_name('pos', browser).send_keys(pos)
        time.sleep(.5)

    find_name('tel', browser).send_keys(tel)
    time.sleep(.5)

    if email != 0:
        find_name('email1', browser).send_keys(email.split('@')[0])
        time.sleep(.5)
        find_name('email2', browser).send_keys(email.split('@')[1])
        
    time.sleep(1)
    find_className('temp_btn2', browser).click()
```

```python
def fourth_page_field(browser, reserach_man, reserach_girl, broad_man, broad_girl, office_nu, pw_text):
    field_data = find_name('a4e1', browser).get_attribute('value')
    field_data1 = find_name('a4e2', browser).get_attribute('value')
    
    man = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table/tbody/tr[2]/td[2]', browser).text
    girl = find_xpath('/html/body/form/center/div/table/tbody/tr/td/table/tbody/tr[2]/td[3]', browser).text
    man_number = man.split(' 명')[0]
    man_number = int(man_number)
    girl_number = girl.split(' 명')[0]
    girl_number = int(girl_number)

    o_and_p_man = man_number - (broad_man + reserach_man)
    o_and_p_girl = girl_number - (broad_girl + reserach_girl)

    office_man = round(o_and_p_man*office_nu/10)
    office_girl = round(o_and_p_girl*office_nu/10)
    
    production_man = o_and_p_man - office_man
    production_girl = o_and_p_girl - office_girl
    
    if reserach_man or reserach_girl:
        fourth_int = 1
    else:
        fourth_int = 0

    # 연구기술직 (남)
    find_name('a4b1', browser).send_keys(reserach_man)
    time.sleep(.5)
    # 연구기술직 (여)
    find_name('a4b2', browser).send_keys(reserach_girl)
    time.sleep(.5)

    # 사무관리작 (남)
    find_name('a4c1', browser).send_keys(office_man)
    time.sleep(.5)
    # 사무관리작 (여)
    find_name('a4c2', browser).send_keys(office_girl)
    time.sleep(.5)

    # 생산직 (남)
    find_name('a4d1', browser).send_keys(production_man)
    time.sleep(.5)
    # 생산직 (여)
    find_name('a4d2', browser).send_keys(production_girl)
    time.sleep(.5)

    # 방송직 (남)
    find_name('a4e1', browser).send_keys(broad_man)
    time.sleep(.5)
    # 방송직 (여)
    find_name('a4e2', browser).send_keys(broad_girl)
    time.sleep(.5)

    try:
        time.sleep(1)
        find_name('q1', browser).send_keys(pw_text)
    except:
        pass
        
    time.sleep(1)
    find_className('temp_btn2', browser).click()
    
    return fourth_int
```

```python
def fiveth_page_field(browser, labeling, r_n, d):
    # SW/SI 개발․설계
    sw = find_name('a5a1', browser)
    # ICT 콘텐츠
    ict_c = find_name('a5a2', browser)
    # 시스템 운영관리
    sy = find_name('a5a3', browser)
    # 정보통신 서비스
    jt = find_name('a5a4', browser)
    # HW 개발․설계
    hw = find_name('a5a5', browser)
    # HW 유지관련
    hw_y = find_name('a5a6', browser)
    # ICT 관련교육
    ict_e = find_name('a5a7', browser)
    # ICT 기술영업
    ict_y = find_name('a5a8', browser)
    
    if labeling in {3, 4, 5, 6, 11, 15, 16}:
        if r_n != 0:
            result = r_n - d
            print("총인원 :", r_n, "명", "/ 개발인원 : ", d, "명", "/ 정_총인원 :", result, "명", "/ 정보통신 :", result, "명")
            jt.send_keys(result)
            time.sleep(.5)
            ### 나머지 값 입력
            ict_c.send_keys(0)
            time.sleep(.5)
            ict_e.send_keys(0)
            time.sleep(.5)
            ict_y.send_keys(0)
            time.sleep(.5)
            sy.send_keys(0)

    # 시스템 운영관리, 정보통신 서비스
    elif labeling in {1, 2, 7, 8, 9, 17}:
        if r_n != 0:
            result = r_n - d
            j = round(result*6/10)
            s = result - round(result*6/10)
            print("총인원 :", r_n, "명", "개발인원 : ", d, "명", "/ 시,정_총인원 : ", result, "명","/ 시스템 :", s, "명", "/ 정보통신 :", j, "명")
            sy.send_keys(s)
            time.sleep(.5)
            jt.send_keys(j)
            time.sleep(.5)

            ### 나머지 값 입력
            ict_c.send_keys(0)
            time.sleep(.5)
            ict_e.send_keys(0)
            time.sleep(.5)
            ict_y.send_keys(0)
            time.sleep(.5)

    # ICT 콘텐츠, 정보통신 서비스, ICT 기술영업
    elif labeling in {10}:
        if r_n != 0:
            result = r_n - d
            j = round(result*7/10)
            i_c = result - round(result*7/10)
            i_y = j - round(j*7/10)
            j = j - i_y
            print("총인원 :", r_n, "명", "개발인원 : ", d, "명", "/ ICT, 정_총인원 :", result, "명","/ ICT_C :", i_c, "명", "/ ICT_E :", i_y, "명", "/ 정보통신 :", j, "명")
            ict_c.send_keys(i_c)
            time.sleep(.5)
            ict_y.send_keys(i_y)
            time.sleep(.5)
            jt.send_keys(j)
            time.sleep(.5)

            ### 나머지 값 입력
            ict_e.send_keys(0)
            time.sleep(.5)
            sy.send_keys(0)
            time.sleep(.5)

    # 시스템 운영관리
    elif labeling in {12, 13, 14}:
        if r_n != 0:
            result = r_n - d
            print("총인원 :", r_n, "명", "개발인원 : ", d, "명", "/ 시_총인원 :", result, "명", "/ 시스템 :", result, "명")
            sy.send_keys(result)
            time.sleep(.5)

            ### 나머지 값 입력
            ict_c.send_keys(0)
            time.sleep(.5)
            ict_y.send_keys(0)
            time.sleep(.5)
            ict_e.send_keys(0)
            time.sleep(.5)
            jt.send_keys(0)
            time.sleep(.5)

    else: print("혹여나 안넣은 숫자 있는지... :", labeling)

    randint = random.randint(6, 7)
    if d != 0:
        sw_number = d
        if d >= 25:
            sw_number = round(d*randint/10)
            hw_total = d - round(d*randint/10)
            if d - round(d*7/10) > 15:
                hw_y_number = round(hw_total - (d - round(d*randint/10))*randint/10)
                hw_d_number = hw_total - hw_y_number
            else:
                hw_d_number = d - round(d*randint/10)
                hw_y_number = 0
        else:
            hw_d_number = 0
            hw_y_number = 0
    else:
        hw_d_number = 0
        hw_y_number = 0
        sw_number = 0

    sw.send_keys(sw_number)
    time.sleep(.5)
    hw.send_keys(hw_d_number)
    time.sleep(.5)
    hw_y.send_keys(hw_y_number)
        
    time.sleep(1)
    find_className('temp_btn2', browser).click()
```
